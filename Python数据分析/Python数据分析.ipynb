## 读写数据

### 读取数据

文件种类:

- pd.read_csv(): 读取csv文件，默认为分隔符为`,`
- pd.read_excel(): 读取excel文件
- pd.read_table(): 读取txt文件，默认分隔符为`\t`

重要参数：

- encoding: 指定编码，默认为`utf-8`
- index_col: 指定一列当索引，默认为None
- usecols: 指定读取哪几列
- parse_dates: 把指定那一列变成datatime(64)ns类型
- date_parser: TODO:
- sheet_name: 指定excel文件的工作表名，默认为`0`
- sep: 指定分隔符
- header: （默认header=0, 即将第一行作为列名），是否有列名，None 没有列名
- header(默认为`infer`): 指定第几行作为列名，默认将第一行作为列名
    - `header=None`: 如果数据没有列名，可使用这个参数

#### python语法中的map

map(function, iterable, ...)

- function: 函数
- iterable: 序列（一个或多个）
- 对序列中的每个值进行函数操作

```python
def square(x):
    return x ** 2
```

```python
_list = [1, 2, 3, 4]
list(map(square, _list)) # map返回的是迭代器，可以通过遍历迭代器查看值或将其转为list
```

```python
# 使用匿名函数
list(map(lambda x: x ** 2, _list))
```

#### Series中的map/apply

Series中的map方法可以接受一个函数或字典型对象

```python
data = pd.DataFrame({'food':['bacon','pulled pork','bacon','Pastrami',
                             'corned beef','Bacon','pastrami','honey ham','nova lox'],
                     'ounces':[4,3,12,6,7.5,8,3,5,6]})
meat_to_animal = {
 'bacon':'pig',
 'pulled pork':'pig',
 'pastrami':'cow',
 'corned beef':'cow',
 'honey ham':'pig',
 'nova lox':'salmon' } 
data
```

```python
data['animal'] = data['food'].map(str.lower).map(meat_to_animal)
data
```

```python
data['animal'].map(dict(pig=0, cow=1)) # 未在字典里的数据，以NaN替代
```

```python
data['animal'].map({'pig': 0, 'cow': 1})
```

## 合并数据

- pd.concat()
- pd.merge()

### pd.concat()

重要参数说明：

- objs：series，dataframe或者是panel构成的序列lsit 
- axis：需要合并链接的轴，0是行，1是列 
- join：连接的方式 inner，或者outer
- ignore_index：忽略需要连接的frame本身的index，当原本的index没有特别意义的时候可以使用，默认为False

#### 合并行

```python
concat_1_10 = senior_score.iloc[0:10]
concat_11_20 = senior_score.iloc[10:20]
concat_21_30 = senior_score.iloc[20:30]
```

```python
type(concat_1_10)
```

```python
pd.concat([concat_1_10, concat_11_20, concat_21_30], axis=0)
```

#### 合并列

```python
concat_math = senior_score.iloc[0:10, 3]
concat_physical = senior_score.iloc[0:10, 5]
concat_chemistry = senior_score.iloc[0:10, 6]
```

```python
pd.concat([concat_math, concat_physical, concat_chemistry], axis=1)
```

### pd.merge()

重要参数说明：

- left与right：两个不同的DataFrame
- how：指的是合并(连接)的方式有inner(内连接),left(左外连接),right(右外连接),outer(全外连接);默认为inner
- on：指的是用于连接的列索引名称。必须存在右右两个DataFrame对象中，如果没有指定且其他参数也未指定则以两个DataFrame的列名交集做为连接键

```python
merge_kh_chinese = senior_score.iloc[0:10, [1, 2]]
merge_kh_math = senior_score.iloc[0:10, [1, 3]]
merge_kh_english = senior_score.iloc[0:10, [1, 4]]
```

```python
pd.merge(merge_kh_chinese, merge_kh_math)
```

```python
pd.merge(merge_kh_chinese, merge_kh_math, on='姓名', left_index=True)
```

## pandas 生成时间序列

常用参数

- `start`: 设置开始时间（可选）
- `end`: 设置结束时间（可选）
- `periods`: 设置时间序列的个数
- `freq`: 设置周期类型

**`freq`**参数列表：

|Alias|Description|
|---|---|
|B|business day frequency|
|C|custom business day frequency|
|D|calendar day frequency|
|W|weekly frequency|
|M|month end frequency|
|SM|semi-month end frequency (15th and end of month)|
|BM|business month end frequency|
|CBM|custom business month end frequency|
|MS|month start frequency|
|SMS|semi-month start frequency (1st and 15th)|
|BMS|business month start frequency|
|CBMS|custom business month start frequency|
|Q|quarter end frequency|
|BQ|business quarter end frequency|
|QS|quarter start frequency|
|BQS|business quarter start frequency|
|A, Y|year end frequency|
|BA, BY|business year end frequency|
|AS, YS|year start frequency|
|BAS, BYS|business year start frequency|
|BH|business hour frequency|
|H|hourly frequency|
|T, min|minutely frequency|
|S|secondly frequency|
|L, ms|milliseconds|
|U, us|microseconds|
|N|nanoseconds|
