# seaborn.boxplot

seaborn.boxplot(*, x=None, y=None, hue=None, data=None, order=None, hue_order=None, orient=None, color=None, palette=None, saturation=0.75, width=0.8, dodge=True, fliersize=5, linewidth=None, whis=1.5, ax=None, **kwargs)

- x: df中的列名或者直接传入series
- y: df中的列名或者直接传入series
- hue: df中的列名, 按照列名中的值分类形成分类的箱线图
- data: df或数组
- orient："v"|"h" 用于控制图像使水平还是竖直显示（这通常是从输入变量的dtype推断出来的，此参数一般当不传入x、y，只传入data的时候使用）
- fliersize：float，用于指示离群值观察的标记大小
- whis：确定离群值的上下界（IQR超过低和高四分位数的比例），此范围之外的点将被识别为异常值。IQR指的是上下四分位的差值。
- width：float，控制箱型图的宽度

## 单变量分析一

```python
# 让可视化更方便更美观
%matplotlib inline
%config InlineBackend.figure_format = 'retina'

from pydataset import data
import seaborn as sns
import matplotlib.pyplot as plt

iris = data('iris')

sns.boxplot(y=iris['Sepal.Length'])  # 直接传入series
plt.show()
```

## 多变量分析

```python
# 让可视化更方便更美观
%matplotlib inline
%config InlineBackend.figure_format = 'retina'

from pydataset import data
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset('tips')

sns.boxplot(x='day', y='total_bill', hue='smoker', data=tips)
plt.show()
```

## 设置whis，离群值边界

```python
# 让可视化更方便更美观
%matplotlib inline
%config InlineBackend.figure_format = 'retina'

from pydataset import data
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset('tips')

sns.boxplot(x='day', y='total_bill', hue='smoker', data=tips, whis=2) # whis设置为2倍IQR
plt.show()
```

## 多列组合

```python
# 让可视化更方便更美观
%matplotlib inline
%config InlineBackend.figure_format = 'retina'

from pydataset import data
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

iris = data('iris')
iris = iris.select_dtypes(include=np.number)

fig = plt.figure(figsize=(4*4, 6))
for i in range(len(iris.columns)):
    plt.subplot(1, 4, i+1)
    sns.boxplot(y=iris.iloc[:, i])
plt.show()
```

```python
# 让可视化更方便更美观
%matplotlib inline
%config InlineBackend.figure_format = 'retina'

from pydataset import data
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

iris = data('iris')
iris = iris.select_dtypes(include=np.number)

fig = plt.figure(figsize=(4*4, 6))
plt.boxplot(x=iris.values, labels=iris.columns)
plt.show()
```
